## ZIPS are in clusters and pops up on click

```python
#popups
#generate data
df = pd.read_csv("zip_code_database.csv")

NY_NJ_CT_df = df[df['state'].isin(['NY','NJ','CT'])]
```

```python
NY_NJ_CT_df.head()
```

```python
pop_county = [i for i in NY_NJ_CT_df['county']]
pop_zips = [i for i in NY_NJ_CT_df['zip']]
pops = list(zip(pop_county,pop_zips))
```

```python
pops
```

```python
data = np.array([NY_NJ_CT_df.latitude,NY_NJ_CT_df.longitude]).T
data
```

```python
data = np.array([NY_NJ_CT_df.latitude,NY_NJ_CT_df.longitude]).T

m = folium.Map([40.7597,-73.9841], zoom_start=5)

plugins.MarkerCluster(data, popups=pops).add_to(m)

plugins.ScrollZoomToggler().add_to(m)

m.save(os.path.join('plugin_maps/pop_up.html'))
m
```

```python
#Boatmarker?
m = folium.Map([40.7597,-73.9841], zoom_start=3)

plugins.BoatMarker(
    location=(42, -70),
    heading=45,
    wind_heading=150,
    wind_speed=45,
    color='#8f8'
).add_to(m)

plugins.BoatMarker(
    location=(45, -75),
    heading=-20,
    wind_heading=46,
    wind_speed=25,
    color='#88f'
).add_to(m)

m.save(os.path.join('plugin_maps/', 'boatmarker.html'))
m
```

```python
#Nice icons
m = folium.Map([45.5, -122], zoom_start=3)

icon_plane = plugins.BeautifyIcon(
    icon='plane',
    border_color='#b3334f',
    text_color='#b3334f')

icon_number = plugins.BeautifyIcon(
    border_color='#00ABDC',
    text_color='#00ABDC',
    number=10,
    inner_icon_style='margin-top:0;')


folium.Marker(
    location=[50, -122],
    popup='British Columbia, Canada',
    icon=icon_number
).add_to(m)

folium.Marker(
    location=[47, -122],
    popup='Portland, OR',
    icon=icon_plane
).add_to(m)


m.save(os.path.join('plugin_maps/', 'nicer_icons.html'))
m
```

```python
m = folium.Map(location=[41.9, -97.3], zoom_start=4)

plugins.Fullscreen(
    position='topright',
    title='Expand me',
    title_cancel='Exit me',
    force_separate_button=True
).add_to(m)

m.save(os.path.join('plugin_maps/', 'screen_control.html'))
m
```

```python
m = folium.Map(location=[0, 0], zoom_start=6)

pop_marker = ['one','two','three',4]

fg = folium.FeatureGroup(name='groups')
m.add_child(fg)

g1 = plugins.FeatureGroupSubGroup(fg, 'group1')
m.add_child(g1)

g2 = plugins.FeatureGroupSubGroup(fg, 'group2')
m.add_child(g2)

folium.Marker([-1, -1], popup=str(pop_marker[0])).add_to(g1)
folium.Marker([1, 1], popup=str(pop_marker[1])).add_to(g1)

folium.Marker([-1, 1]).add_to(g2)
folium.Marker([1, -1]).add_to(g2)

folium.LayerControl(collapsed=False).add_to(m)
m.save(os.path.join('plugin_maps/', 'feature_groups.html'))
m
```

```python
m = folium.Map(location=(30, 20), zoom_start=4)

minimap = plugins.MiniMap()
m.add_child(minimap)

m.save(os.path.join('plugin_maps/', 'mini_map.html'))
m
```

```python
m = plugins.DualMap(location=(52.1, 5.1), tiles=None, zoom_start=8)

folium.TileLayer('cartodbpositron').add_to(m.m2)
folium.TileLayer('openstreetmap').add_to(m)

fg_both = folium.FeatureGroup(name='markers_both').add_to(m)
fg_1 = folium.FeatureGroup(name='markers_1').add_to(m.m1)
fg_2 = folium.FeatureGroup(name='markers_2').add_to(m.m2)

icon_red = folium.Icon(color='red')
folium.Marker((52, 5), tooltip='both', icon=icon_red).add_to(fg_both)
folium.Marker((52.4, 5), tooltip='left').add_to(fg_1)
folium.Marker((52, 5.4), tooltip='right').add_to(fg_2)

folium.LayerControl(collapsed=False).add_to(m)
m.save(os.path.join('plugin_maps/', 'dual_map.html'))
m
```
